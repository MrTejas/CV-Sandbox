```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import time
import numpy as np
import matplotlib.pyplot as plt
import os    
import wandb 
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

%matplotlib inline
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

```python
# seeds for reproducibility
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)
```

# 1. Class for LeNet 

Following is the Architecture used for LeNet.


- Input Image `28,28,1`

- Image to  C1     : `1,5,5,6` convolution layer
- C1    to  S2     : `6,2,2,6` Max Pooling layer
- S2    to  C3     : `6,5,5,16` convolution layer
- C3    to  S4     : `16,2,2,16` Max Pooling layer
- S4    to  C5     : `16*4*4, 120` Linear Layer + ReLU
- C5    to  F6     : `120, 84` Linear Layer + ReLU
- F6    to  Output : `84, 10` Linear Layer+ReLU



```python

class MNISTTrainer:
    def __init__(self, batch_size=64, learning_rate=0.001, optimizer='adam'):
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.optimizer_name = optimizer

        # configuring and initializing wandb for logging the run
        config = {"learning_rate": self.learning_rate, "batch_size": self.batch_size, "optimizer": self.optimizer_name}
        wandb.init(project="mnist-training",config=config)  # Initialize wandb
        
        # initializing the dataset
        self.initialize_data()
        
        # NN modulefor LeNet architecture
        class LeNet(nn.Module):
            def __init__(self):
                super(LeNet, self).__init__()
                self.conv1 = nn.Conv2d(1, 6, kernel_size=5)
                self.conv2 = nn.Conv2d(6, 16, kernel_size=5)
                self.fc1 = nn.Linear(16*4*4, 120)
                self.fc2 = nn.Linear(120, 84)
                self.fc3 = nn.Linear(84, 10)

            def forward(self, x):
                x = torch.relu(self.conv1(x))
                x = torch.max_pool2d(x, 2)
                x = torch.relu(self.conv2(x))
                x = torch.max_pool2d(x, 2)
                x = x.view(-1, 16*4*4)
                x = torch.relu(self.fc1(x))
                x = torch.relu(self.fc2(x))
                x = self.fc3(x)
                return x

        # if gpu is available, moving the model to gpu
        self.net = LeNet()
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.net.to(self.device)

        # Initialize the model, loss function, and optimizer
        self.criterion = nn.CrossEntropyLoss()
        if optimizer == 'adam':
            self.optimizer = optim.Adam(self.net.parameters(), lr=self.learning_rate)
        elif optimizer == 'sgd':
            self.optimizer = optim.SGD(self.net.parameters(), lr=self.learning_rate, momentum=0.9)
        elif optimizer == 'rmsprop':
            self.optimizer = optim.RMSprop(self.net.parameters(), lr=self.learning_rate)
        else:
            raise ValueError("Optimizer not found... Should be 'adam', 'sgd', or 'rmsprop'")

        self.train_losses = []
        self.test_losses = []
        self.test_accuracies = []
        self.epoch_times = []
        
    def initialize_data(self):  
        
        # defining transform to (a) convert the dataset to tensor (b)Normalize with 0.5 mean, 0.5 std
        self.transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))
        ])

        # downloading and loading the dataset
        self.trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=self.transform)
        self.testset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=self.transform)
        self.trainloader = torch.utils.data.DataLoader(self.trainset, batch_size=self.batch_size, shuffle=True)
        self.testloader = torch.utils.data.DataLoader(self.testset, batch_size=self.batch_size, shuffle=False)

        
    # 1 training epoch iteration
    def train_epoch(self):
        self.net.train()
        running_loss = 0.0
        start_time = time.time()
        for inputs, labels in self.trainloader:
            inputs, labels = inputs.to(self.device), labels.to(self.device)
            self.optimizer.zero_grad()
            outputs = self.net(inputs)
            loss = self.criterion(outputs, labels)
            loss.backward()
            self.optimizer.step()
            running_loss += loss.item() * inputs.size(0)
        epoch_loss = running_loss / len(self.trainloader.dataset)
        end_time = time.time()
        epoch_time = end_time - start_time
        return epoch_loss, epoch_time

    # 1 evaluation epoch iteration
    def eval_epoch(self):
        self.net.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in self.testloader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                outputs = self.net(inputs)
                loss = self.criterion(outputs, labels)
                running_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        epoch_loss = running_loss / len(self.testloader.dataset)
        accuracy = correct / total
        return epoch_loss, accuracy

    # function to train the model
    def train_model(self, num_epochs=10):
        for epoch in range(num_epochs):
            train_loss, epoch_time = self.train_epoch()
            test_loss, test_accuracy = self.eval_epoch()
            self.train_losses.append(train_loss)
            self.test_losses.append(test_loss)
            self.test_accuracies.append(test_accuracy)
            self.epoch_times.append(epoch_time)
            
            # Log metrics to wandb
            wandb.log({"Train Loss": train_loss, "Test Loss": test_loss, "Test Accuracy": test_accuracy, "Epoch Time": epoch_time})  # Log metrics
            
            print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}, Epoch Time: {epoch_time:.2f} seconds")

    # function to plot the training and test losses
    def plot_losses(self):
        plt.figure(figsize=(10, 5))
        plt.plot(range(1, len(self.train_losses) + 1), self.train_losses, label='Train Loss')
        plt.plot(range(1, len(self.test_losses) + 1), self.test_losses, label='Test Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Training and Test Losses')
        plt.legend()
        plt.grid(True)
        plt.show()
    
    # plot epoch time
    def plot_epoch_times(self):
        plt.figure(figsize=(10, 5))
        plt.plot(range(1, len(self.epoch_times) + 1), self.epoch_times, marker='o')
        plt.xlabel('Epoch')
        plt.ylabel('Time (seconds)')
        plt.title('Epoch Time')
        plt.grid(True)
        plt.show()

```

### Example Usage of the class (for the best hyperparameter setting)

```python
# # Example usage:
batch_size = 40
optimizer = 'adam'
learning_rate = 0.001
num_epoches = 15

trainer = MNISTTrainer(batch_size=batch_size, learning_rate=learning_rate, optimizer=optimizer)
trainer.train_model(num_epochs=num_epoches)
trainer.plot_losses()
trainer.plot_epoch_times()
```

## Trends in Classification Accuracy

Since we have kept the number of epoches constant (10), we can compare the importance and correlation of the 3 hyperparameters, `batch_size`, `learning_rate` and `optimizer` with the `test_accuracy`.



The above image shows the correlation of the hyperparameters with the `test_accuracy`. Following observations can be made :-

- **Batch Size :** We can observe that `batch_size` has a positive correlation with accuracy, meaning higher batch size results in higher accuracy. However, this result cannot be generalized for all ranges of values for batch sizes. Since we have run the model for hyperparameters having batch sizes moderately smaller than the optimum. Larger batch sizes will obviously lead to faster convergence and thus require lesser number of epoches (10 here), however, excessively large batch sizes might lead to poor generalization of the dataset and converge to a suboptimal solution.

---
- **Learning Rate :** It determines the size of the steps taken while optimization of the model to update model parameters. A large step size results in vigorous updates and may lead to oscillations or divergence. However, a very small learning rate causes very slow convergence and thus require much more iterations (number of epoches) to converge. In general, a smaller learning rate is good for ensuring that we reach the global optimum and is used as long as the model has enough number of epoches as shown by the large negative correlation in the above figure. In our model, for 10 epoches, a learning rate of around `0.001` works well in most cases. 
---
- **Optimizer :** The choice between these optimizers depends on factors such as the dataset size, model architecture, and computational resources. `Adam` and `rmsprop` tend to converge faster, `SGD` helps in better generalization to unseen (test) data and `momentum` (0.9 for our model) with `SGD` prevents it from converging to local optimim.  In our case, `Adam` performs better than other optimizers. However, the correlations in oprimizers is lower than those in batch size and learning rate.
---




---
Overall, while `batch_size` shows a positive correlation with test accuracy, we observe a negative correlation with `learning_rate`. By importance, `SGD optimizer` shows largest importance followed by `batch_size` and `learning_rate`. The following parallel-coordinates visualization well represents how the hyperparameters affect the model performance.




# 2. Deep LeNet Architecture having double the layers in LeNet



To double the number of layers in the original LeNet architecture, we introduce 2 extra `conv2d` and `max_pool` layers. We also changed our architecture to add padding in the convolutional layers to keep the size of the patches reasonable large (The network will learn almost nothing from 1,1 images)

```python

class DeepMNISTTrainer:
    def __init__(self, batch_size=64, learning_rate=0.001, optimizer='adam'):
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.optimizer_name = optimizer

        # initializing the dataset
        self.initialize_data()
        
        # NN modulefor LeNet architecture
        class LeNet(nn.Module):
            def __init__(self):
                super(LeNet, self).__init__()
                self.conv1 = nn.Conv2d(1, 6, kernel_size=5)
                self.conv2 = nn.Conv2d(6, 16, kernel_size=5, padding=2)
                self.conv3 = nn.Conv2d(16, 20, kernel_size=5, padding=2)
                self.conv4 = nn.Conv2d(20, 24, kernel_size=3, padding=1)
                self.fc1 = nn.Linear(24*2*2, 120)
                self.fc2 = nn.Linear(120, 84)
                self.fc3 = nn.Linear(84, 10)

            def forward(self, x):
                x = torch.relu(self.conv1(x))
                x = torch.max_pool2d(x, 2)
                x = torch.relu(self.conv2(x))
                x = torch.max_pool2d(x, 2)
                x = torch.relu(self.conv3(x))
                x = torch.max_pool2d(x, 2)
                x = torch.relu(self.conv4(x))
                x = torch.max_pool2d(input=x, kernel_size=2, stride=1)
                x = x.view(-1, 24*2*2)
                x = torch.relu(self.fc1(x))
                x = torch.relu(self.fc2(x))
                x = self.fc3(x)
                return x

        # if gpu is available, moving the model to gpu
        self.net = LeNet()
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.net.to(self.device)

        # Initialize the model, loss function, and optimizer
        self.criterion = nn.CrossEntropyLoss()
        if optimizer == 'adam':
            self.optimizer = optim.Adam(self.net.parameters(), lr=self.learning_rate)
        elif optimizer == 'sgd':
            self.optimizer = optim.SGD(self.net.parameters(), lr=self.learning_rate, momentum=0.9)
        elif optimizer == 'rmsprop':
            self.optimizer = optim.RMSprop(self.net.parameters(), lr=self.learning_rate)
        else:
            raise ValueError("Optimizer not found... Should be 'adam', 'sgd', or 'rmsprop'")

        self.train_losses = []
        self.test_losses = []
        self.test_accuracies = []
        self.epoch_times = []
        
    def initialize_data(self):  
        
        # defining transform to (a) convert the dataset to tensor (b)Normalize with 0.5 mean, 0.5 std
        self.transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))
        ])

        # downloading and loading the dataset
        self.trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=self.transform)
        self.testset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=self.transform)
        self.trainloader = torch.utils.data.DataLoader(self.trainset, batch_size=self.batch_size, shuffle=True)
        self.testloader = torch.utils.data.DataLoader(self.testset, batch_size=self.batch_size, shuffle=False)

        
    # 1 training epoch iteration
    def train_epoch(self):
        self.net.train()
        running_loss = 0.0
        start_time = time.time()
        for inputs, labels in self.trainloader:
            inputs, labels = inputs.to(self.device), labels.to(self.device)
            self.optimizer.zero_grad()
            outputs = self.net(inputs)
            loss = self.criterion(outputs, labels)
            loss.backward()
            self.optimizer.step()
            running_loss += loss.item() * inputs.size(0)
        epoch_loss = running_loss / len(self.trainloader.dataset)
        end_time = time.time()
        epoch_time = end_time - start_time
        return epoch_loss, epoch_time

    # 1 evaluation epoch iteration
    def eval_epoch(self):
        self.net.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in self.testloader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                outputs = self.net(inputs)
                loss = self.criterion(outputs, labels)
                running_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        epoch_loss = running_loss / len(self.testloader.dataset)
        accuracy = correct / total
        return epoch_loss, accuracy

    # function to train the model
    def train_model(self, num_epochs=10):
        for epoch in range(num_epochs):
            train_loss, epoch_time = self.train_epoch()
            test_loss, test_accuracy = self.eval_epoch()
            self.train_losses.append(train_loss)
            self.test_losses.append(test_loss)
            self.test_accuracies.append(test_accuracy)
            self.epoch_times.append(epoch_time)
                        
            print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}, Epoch Time: {epoch_time:.2f} seconds")

    # function to plot the training and test losses
    def plot_losses(self):
        plt.figure(figsize=(10, 5))
        plt.plot(range(1, len(self.train_losses) + 1), self.train_losses, label='Train Loss')
        plt.plot(range(1, len(self.test_losses) + 1), self.test_losses, label='Test Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Training and Test Losses')
        plt.legend()
        plt.grid(True)
        plt.show()
    
    # plot epoch time
    def plot_epoch_times(self):
        plt.figure(figsize=(10, 5))
        plt.plot(range(1, len(self.epoch_times) + 1), self.epoch_times, marker='o')
        plt.xlabel('Epoch')
        plt.ylabel('Time (seconds)')
        plt.title('Epoch Time')
        plt.grid(True)
        plt.show()

```


## Performance for double the layers in the network

What happens when layers in the Network are doubled?

- **More Complexity :** More the number of layers, more complex the model is. While this might help the model to learn more complex dataset and more non-linear decision boundaries, it is very possible that it might lead to overfitting and eventually lower test accuracy.
---
- **Increased Parameter Count :** Double the number of layers means more parameter count and thus, more trainable parameters. This clearly means higher train-time and inference-time of the model. 
---
- **Saturated Performance :** We already see that for our current model only, the accuracy reaches as high as **99 %** and thus, more layers in the model may increase the accuracy, but probably by negligible amount which does not compensate for the high train-time and inference time of the model.
---

The performance of the Deep LeNet with double the number of layers drops from 99.23% accuracy of the original LeNet to 98.92%. This drop is however quite small and cannot be explained based on only the depth of the network. It is quite possible that hyperparameter tuning fot this architecture can increase this accuracy. However, the small yet significant decrease in performance(accuracy) can be explained in terms of the model being "too complicated" for the dataset. Increasing the number of layers in the architecture increases the complexity of the network, resulting in the model overfitting to the training set, while performing poorly (however 98.92% is not poor) on the test-set. ```python

```python
class MyMNISTTrainer:
    def __init__(self, batch_size=64, learning_rate=0.001, optimizer='adam', trainset=None, testset=None, trainloader=None, testloader=None):
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.optimizer_name = optimizer
        
        self.num_train_samples = len(trainloader)*self.batch_size

        # configuring and initializing wandb for logging the run
        config = {"learning_rate": self.learning_rate, "batch_size": self.batch_size, "optimizer": self.optimizer_name, "#Train Samples":self.num_train_samples}
        wandb.init(project="my-mnist-training",config=config)  # Initialize wandb
        
        # initializing the dataset
        self.trainset = trainset
        self.testset = testset
        self.trainloader = trainloader
        self.testloader = testloader
        
        # NN modulefor LeNet architecture
        class LeNet(nn.Module):
            def __init__(self):
                super(LeNet, self).__init__()
                self.conv1 = nn.Conv2d(1, 6, kernel_size=5)
                self.conv2 = nn.Conv2d(6, 16, kernel_size=5)
                self.fc1 = nn.Linear(16*4*4, 120)
                self.fc2 = nn.Linear(120, 84)
                self.fc3 = nn.Linear(84, 10)

            def forward(self, x):
                x = torch.relu(self.conv1(x))
                x = torch.max_pool2d(x, 2)
                x = torch.relu(self.conv2(x))
                x = torch.max_pool2d(x, 2)
                x = x.view(-1, 16*4*4)
                x = torch.relu(self.fc1(x))
                x = torch.relu(self.fc2(x))
                x = self.fc3(x)
                return x

        # if gpu is available, moving the model to gpu
        self.net = LeNet()
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.net.to(self.device)

        # Initialize the model, loss function, and optimizer
        self.criterion = nn.CrossEntropyLoss()
        if optimizer == 'adam':
            self.optimizer = optim.Adam(self.net.parameters(), lr=self.learning_rate)
        elif optimizer == 'sgd':
            self.optimizer = optim.SGD(self.net.parameters(), lr=self.learning_rate, momentum=0.9)
        elif optimizer == 'rmsprop':
            self.optimizer = optim.RMSprop(self.net.parameters(), lr=self.learning_rate)
        else:
            raise ValueError("Optimizer not found... Should be 'adam', 'sgd', or 'rmsprop'")

        self.train_losses = []
        self.test_losses = []
        self.test_accuracies = []
        self.epoch_times = []
        
        
    # 1 training epoch iteration
    def train_epoch(self):
        self.net.train()
        running_loss = 0.0
        start_time = time.time()
        for inputs, labels in self.trainloader:
            inputs, labels = inputs.to(self.device), labels.to(self.device)
            self.optimizer.zero_grad()
            outputs = self.net(inputs)
            loss = self.criterion(outputs, labels)
            loss.backward()
            self.optimizer.step()
            running_loss += loss.item() * inputs.size(0)
        epoch_loss = running_loss / len(self.trainloader.dataset)
        end_time = time.time()
        epoch_time = end_time - start_time
        return epoch_loss, epoch_time

    # 1 evaluation epoch iteration
    def eval_epoch(self):
        self.net.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in self.testloader:
                inputs, labels = inputs.to(self.device), labels.to(self.device)
                outputs = self.net(inputs)
                loss = self.criterion(outputs, labels)
                running_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        epoch_loss = running_loss / len(self.testloader.dataset)
        accuracy = correct / total
        return epoch_loss, accuracy

    # function to train the model
    def train_model(self, num_epochs=10):
        for epoch in range(num_epochs):
            train_loss, epoch_time = self.train_epoch()
            test_loss, test_accuracy = self.eval_epoch()
            self.train_losses.append(train_loss)
            self.test_losses.append(test_loss)
            self.test_accuracies.append(test_accuracy)
            self.epoch_times.append(epoch_time)
            
            # Log metrics to wandb
            wandb.log({"Train Loss": train_loss, "Test Loss": test_loss, "Test Accuracy": test_accuracy, "Epoch Time": epoch_time})  # Log metrics
            
            print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}, Epoch Time: {epoch_time:.2f} seconds")

    # function to plot the training and test losses
    def plot_losses(self):
        plt.figure(figsize=(10, 5))
        plt.plot(range(1, len(self.train_losses) + 1), self.train_losses, label='Train Loss')
        plt.plot(range(1, len(self.test_losses) + 1), self.test_losses, label='Test Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Training and Test Losses')
        plt.legend()
        plt.grid(True)
        plt.show()
    
    # plot epoch time
    def plot_epoch_times(self):
        plt.figure(figsize=(10, 5))
        plt.plot(range(1, len(self.epoch_times) + 1), self.epoch_times, marker='o')
        plt.xlabel('Epoch')
        plt.ylabel('Time (seconds)')
        plt.title('Epoch Time')
        plt.grid(True)
        plt.show()

```

# 6. Two-Layer Transformer Encoder

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

trainset = datasets.MNIST('~/.pytorch/MNIST_data/', train=True, download=True, transform=transform)
testset = datasets.MNIST('~/.pytorch/MNIST_data/', train=False, download=True, transform=transform)
```



Transformer encoder layer architecture (left) and schematic overview of a multi-head self-attention block (right).


```python



class TransformerEncoderLayer(nn.Module):
    def __init__(self, d_model, nhead, dim_feedforward=2048, dropout=0.1):
        # inheriting the pytorch transformer encoder layer
        super(TransformerEncoderLayer, self).__init__()
        
        # initializing the encoder with layers and attention heads
        self.self_attn = nn.MultiheadAttention(d_model, nhead, dropout=dropout)
        
        # linear layer for actual prediction
        self.linear1 = nn.Linear(d_model, dim_feedforward)
        self.linear2 = nn.Linear(dim_feedforward, d_model)

        # dropout layers with some dropout probability for regularization
        self.dropout = nn.Dropout(dropout)
        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)
        
        # Layer Normalization
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, src, src_mask=None):
        src2 = self.self_attn(src, src, src, attn_mask=src_mask)[0]
        src = src + self.dropout1(src2)
        src = self.norm1(src)
        src2 = self.linear2(self.dropout(torch.relu(self.linear1(src))))
        src = src + self.dropout2(src2)
        src = self.norm2(src)
        return src

class TransformerEncoder(nn.Module):
    def __init__(self, num_layers, d_model, nhead, dim_feedforward=2048, dropout=0.1):
        super(TransformerEncoder, self).__init__()
        encoder_layers = nn.TransformerEncoderLayer(d_model, nhead, dim_feedforward, dropout,batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers)

    def forward(self, src, src_mask=None):
        output = self.transformer_encoder(src, src_mask)
        return output

class ViTClassifier(nn.Module):
    def __init__(self, num_layers, d_model, nhead, num_classes, dim_feedforward=2048, dropout=0.1):
        super(ViTClassifier, self).__init__()
        
        # embedding layer to convert the (28,28) images to d_model dim.
        self.embedding = nn.Linear(28 * 28, 784 * d_model) 
        # layer to add positional encoding to the images
        self.pos_encoder = nn.Parameter(torch.randn(1, 784, d_model))
        # creating the transformer
        self.transformer_encoder = TransformerEncoder(num_layers, d_model, nhead, dim_feedforward, dropout)
        # to map the transformer output to number of classes
        self.fc = nn.Linear(d_model, num_classes)

        
        
    def forward(self, x):
        x = x.view(x.size(0), -1)  # Flatten input images
        x = self.embedding(x)
        x = x.view(x.size(0), 784, -1)  # Reshape to [batch_size, 784, d_model]
        x = x + self.pos_encoder
        x = self.transformer_encoder(x)  # Transformer encoding
        x = x.mean(dim=1)  # Average pooling
        x = self.fc(x)  # Classification
        return x


def train(model, trainloader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    for inputs, labels in trainloader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    return running_loss / len(trainloader)

def evaluate(model, dataloader, device):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    return correct / total
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize parameters
num_layers = 2
d_model = 64
nhead = 4
num_classes = 10
batch_size = 64


batch_size_6000 = 64  # For 6000 training samples
batch_size_60000 = 64  # For 60000 training samples
num_workers = 2  # Number of workers for data loading (for fast loading)

# Calculate the number of batches needed to cover the desired number of samples
num_batches_6000 = 6000 // batch_size_6000
num_batches_60000 = 60000 // batch_size_60000


# For 6000 training samples
trainloader_6000 = torch.utils.data.DataLoader(
    trainset,
    batch_size=batch_size_6000,
    num_workers=num_workers,
    sampler=torch.utils.data.SubsetRandomSampler(range(6000))
)

# For 60000 training samples
trainloader_60000 = torch.utils.data.DataLoader(
    trainset,
    batch_size=batch_size_60000,
    num_workers=num_workers,
    shuffle=True  # Shuffle the entire dataset
)

print("Number of train samples for 6K loader  : ",len(trainloader_6000)*batch_size_6000)
print("Number of train samples for 60K loader : ",len(trainloader_60000)*batch_size_60000)

testloader = DataLoader(testset, batch_size=batch_size)

model_6k = ViTClassifier(num_layers, d_model, nhead, num_classes).to(device)
model_60k = ViTClassifier(num_layers, d_model, nhead, num_classes).to(device)

criterion = nn.CrossEntropyLoss()
optimizer_6k = optim.Adam(model_6k.parameters(), lr=0.001)
optimizer_60k = optim.Adam(model_60k.parameters(), lr=0.001)
```

```python

# Train and evaluate with 6K images
for epoch in range(5):
    train_loss = train(model_6k, trainloader_6000, criterion, optimizer_6k, device)
    accuracy = evaluate(model_6k, testloader, device)
    print(f"6K Images: Epoch {epoch+1}, Loss: {train_loss:.4f}, Accuracy: {accuracy:.4f}")
```
